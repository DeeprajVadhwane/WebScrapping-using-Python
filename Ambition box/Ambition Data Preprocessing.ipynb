### Renaming the Columns

```python
df = df.rename(columns={'name1': 'company_name', 'data_1': 'year_of_started', 'data_2': 'sector', 'data_5': 'location', 'data_7' : 'nationalality', 'salary1': 'Average_salary', 'salary3': 'min_max', 'data_6':'founder','Container_1':'no_reviews', 'Container_3':'no_interviews', 'Container_4':'jobs'})
```

```python
df.head()
```

```python
df.columns
```

```python
columns_to_drop = ['data_3', 'data_4', 'Container_2', 'Container_5', 'Container_6', 'Container_7',
    'Container_8', 'name2', 'name3', 'name4', 'data_8', 'data_9']
```

```python
df = df.drop(columns=columns_to_drop)
```

```python
df.head()
```

```python
df = df.drop(columns='salary2')
```

```python
df.head()
```

```python
r = []
for i in df['year_of_started']:
    #print(i)
    r.append(i.split('(')[0])
r
```

```python
df['year_of_started'] = r
```

```python
df.head()
```

```python
df.tail()
```

```python
df.info()
```

```python
df['year_of_started'] = df['year_of_started'].astype('int')
```

```python
df.loc[df['no_reviews'].str.contains('k'), 'no_reviews'] = df.loc[df['no_reviews'].str.contains('k'), 'no_reviews'].str.replace('k', '').astype(float) * 1000

print(df['no_reviews'])
```

```python
df.loc[df['no_interviews'].str.contains('k'), 'no_interviews'] = df.loc[df['no_interviews'].str.contains('k'), 'no_interviews'].str.replace('k', '').astype(float) * 1000

print(df['no_interviews'])
```

```python
df.head()
```

```python
df.loc[df['jobs'].str.contains('k'), 'jobs'] = df.loc[df['jobs'].str.contains('k'), 'jobs'].str.replace('k', '').astype(float) * 1000

print(df['jobs'])
```

```python
df.head()
```

```python
df['Average_salary'] = df['Average_salary'].str.replace(r'[₹\n\t/]', '', regex=True).str.replace(',', '', regex=False).str.extract(r'(\d+)').astype(int)
```

```python
df.head()
```

```python
df.tail()
```

```python
df.info()
```

```python
df['no_interviews'] = df['no_interviews'].str.replace(',', '').astype(float)
```

```python
df.isnull().sum()
```

```python
df['jobs'].fillna(df['jobs'].median(),inplace=True)
```

```python
df['no_interviews'].fillna(df['no_interviews'].median(),inplace=True)
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df[['min_salary','max_salary']]=[col.replace('L/yr',"").replace("₹","").replace("(","").replace(")","").split("-") for col in df['min_max']]
```

```python
df['min_salary']=df['min_salary'].astype('float64')
df['max_salary']=df['max_salary'].astype('float64')
```

```python
df.drop('min_max',axis=1,inplace=True)
```
